<h6><font color="BlueViolet">시리즈에 값 쓰기</font></h6>

<h3><font color="red">시리즈 연산</font></h3>

```python
myindex31 = ['에스프레소','아메리카노','마끼아또','바닐라라떼']
mylist31 = [30,40,50,60]

myindex32 = ['에스프레소','아메리카노','마끼아또','콜드브루']
mylist32 = [30,40,50,70]

myseries31 = pd.Series(data=mylist31,index=myindex31)

myseries32 = pd.Series(data=mylist32,index=myindex32)
```

```python
myseries31
```

```python
myseries32
```

```python
su = 10
print(myseries31 + su)
```

```python
print(myseries31.add(su))
```

```python
print(myseries31.sub(su))
```

```python
print(myseries31.mul(su))
```

```python
mylist04 = ['카페모카','에스프레소','에스프레소','카페라떼','에스프레소','아메리카노','아메리카노','에스프레소','카페모카','카페모카']
myseries04 = pd.Series(mylist04)
myseries04
```

```python
myunique04 = myseries04.unique() # DB의 Distinct와 유사(중복제거)
myunique04
```

```python
mycount04 = myseries04.count()
mycount04
```

```python
myseries04.value_counts() # 빈도수 파악해서 역순으로 정렬
```

```python
myseries04.value_counts(normalize=True) # 빈도수를 비율로 보여줌
```

```python
myindex05 =[ '아메리카노','에스프레소','카페라떼','카페모카','콜드브루','비날라라떼','마끼아또']
mylist05 = [5000,3000,2000,6000,4000, 8000,7000]
myseries05 = pd.Series(data = mylist05, index = myindex05)
myseries05
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
myseries05.plot(kind='line', grid = True, rot=20)
plt.xlabel('상품명',fontsize=14)
plt.ylabel('단가',fontsize=14)
plt.title('품목별 단가 꺽은 선',fontsize=18)
```

```python
myseries05.plot(kind='bar', grid = True, rot=0, color=['r','g','b','y','m'])
plt.xlabel('상품명',fontsize=14)
plt.ylabel('단가',fontsize=14)
plt.title('품목별 단가 꺽은 선',fontsize=18)
```
